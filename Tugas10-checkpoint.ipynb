    <li>Gender</li>
    <li>Jurusan</li>
    <li>Organisasi</li>
    <li>Tingkat-Aktif</li>
    <li>Semester</li>
    <li>Nilai Etika</li>
    <li>Nilai Soft Skill</li>
    <li>Kontribusi</li>
    <li>Penghargaan</li>
</ol>

```python
import numpy as np
import pandas as pd

n_data = 1000

# Data acak
usia = np.random.randint(18, 25, size=n_data)
gender = np.random.choice(['L', 'P'], size=n_data)
jurusan = np.random.choice(['TI', 'TS', 'TE', 'TM'], size=n_data)
organisasi = np.random.choice(['BEM', 'HIMA', 'UKM Musik', 'UKM Olahraga', 'Tidak Aktif'], size=n_data)
tingkat_aktif = np.random.choice(['Sangat Aktif', 'Aktif', 'Kurang Aktif', 'Pasif'], size=n_data)
semester = np.random.randint(1, 9, size=n_data)
nilaiEtika = np.round(np.random.uniform(2.0, 4.0, size=n_data), 2)
nilaiSoftskill = np.round(np.random.uniform(2.0, 4.0, size=n_data), 2)
kontribusi = np.random.choice(['Tinggi', 'Sedang', 'Rendah'], size=n_data)
penghargaan = np.random.choice(['Ya', 'Tidak'], size=n_data)

# Buat DataFrame
df = pd.DataFrame({
    'Usia': usia,
    'Gender': gender,
    'Jurusan': jurusan,
    'Organisasi': organisasi,
    'Tingkat Aktif': tingkat_aktif,
    'Semester': semester,
    'Nilai Etika': nilaiEtika,
    'Nilai Softskill': nilaiSoftskill,
    'Kontribusi': kontribusi,
    'Penghargaan': penghargaan
})

# Analisis dasar
print(df.info())
print(f"Usia termuda: {df.Usia.min()}")
print(f"Usia tertua: {df.Usia.max()}")
print(f"Jumlah mahasiswa laki-laki: {len(df[df['Gender'] == 'L'])}")
print(f"Jumlah mahasiswa perempuan: {len(df[df['Gender'] == 'P'])}")
print(f"Jumlah mahasiswa tiap jurusan:\n{df['Jurusan'].value_counts()}")
print(f"Jumlah mahasiswa tiap organisasi:\n{df['Organisasi'].value_counts()}")
print(f"Jumlah mahasiswa dengan tingkat aktif:\n{df['Tingkat Aktif'].value_counts()}")
print(f"Rata-rata nilai etika: {df['Nilai Etika'].mean():.2f}")
print(f"Rata-rata nilai softskill: {df['Nilai Softskill'].mean():.2f}")
print(f"Jumlah kontribusi:\n{df['Kontribusi'].value_counts()}")
print(f"Jumlah yang mendapat penghargaan: {len(df[df['Penghargaan'] == 'Ya'])}")
print(f"Jumlah semester akhir (>= 7): {len(df[df['Semester'] >= 7])}")

# Simpan ke CSV
df.to_csv('dataMahasiswaEkstrakurikuler.csv', index=False)
```
